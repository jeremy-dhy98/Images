```python
from PIL import Image, ImageDraw, ImageFont
import IPython.display as ipd
import cairo
import math
```

```python
# Create a blank canvas
width, height = 400, 400
image = Image.new("RGB", (width, height), "white")
draw = ImageDraw.Draw(image)

# Draw a circle
draw.ellipse((100, 100, 300, 300), fill="blue", outline="black", width=5)

# Add text
font = ImageFont.truetype("arial.ttf", size=30)  
text = "My Logo"

# Calculate text size using textbbox
text_bbox = draw.textbbox((0, 0), text, font=font)
text_width = text_bbox[2] - text_bbox[0]
text_height = text_bbox[3] - text_bbox[1]

# Calculate the center position
text_x = (width - text_width) / 2
text_y = (height - text_height) / 2
# text_x = (width - text_width) // 2
# text_y = (height - text_height) // 2
draw.text((text_x, text_y), text, fill="white", font=font)

# Save the image
image.save("logo.png")
ipd.Image("logo.png", width=350, height=350, embed=True)

```

```python
# Canvas size
WIDTH, HEIGHT = 500, 500

# Create the surface and context
surface = cairo.ImageSurface(cairo.Format.ARGB32, WIDTH, HEIGHT)
context = cairo.Context(surface)

# Background color
context.rectangle(0, 0, WIDTH, HEIGHT)
context.set_source_rgb(1, 1, 1)  # White background
context.fill()

# Draw the outer circle
circle_center_x = WIDTH / 2
circle_center_y = HEIGHT / 2
circle_radius = 200

context.arc(circle_center_x, circle_center_y, circle_radius, 0, 2 * math.pi)
context.set_source_rgb(0.1, 0.6, 0.8)  # Light blue
context.fill_preserve()  # Fill but keep path for stroke
context.set_source_rgb(0, 0, 0)  # Black outline
context.set_line_width(2)
context.stroke()

# Draw the star
def draw_star(context, cx, cy, spikes, outer_radius, inner_radius):
    angle = math.pi / spikes
    context.move_to(cx, cy - outer_radius)  # Starting point
    for i in range(2 * spikes + 1):
        radius = outer_radius if i % 2 == 0 else inner_radius
        x = cx + math.sin(i * angle) * radius
        y = cy - math.cos(i * angle) * radius
        context.line_to(x, y)
    context.close_path()

draw_star(context, circle_center_x, circle_center_y, spikes=5, outer_radius=150, inner_radius=75)
context.set_source_rgb(1, 1, 0)  # Yellow fill
context.fill_preserve()
context.set_source_rgb(0, 0, 0)  # Black outline
context.stroke()

# Add text "Jeremy" inside the star
context.select_font_face("Arial", cairo.FONT_SLANT_NORMAL, cairo.FONT_WEIGHT_BOLD)
context.set_font_size(40)
text = "Jeremy"
x_bearing, y_bearing, text_width, text_height, _, _ = context.text_extents(text)
context.move_to(circle_center_x - text_width / 2, circle_center_y + text_height / 4)
context.set_source_rgb(0, 0, 0)  # Black text
context.show_text(text)

# Circle that touches the edges of the star
outer_radius = 150  # Match the star's outer radius
context.arc(WIDTH / 2, HEIGHT / 2, outer_radius, 0, 2 * math.pi)
context.set_source_rgb(1, 1, 1)  # White outline
context.set_line_width(3)  # Thin stroke
context.stroke()

# Add "DEV" at the bottom of the star
context.set_font_size(20)
dev_text = "DEV"
x_bearing, y_bearing, text_width, text_height, _, _ = context.text_extents(dev_text)
context.move_to(circle_center_x - text_width / 2, circle_center_y + 80)
context.set_source_rgb(0.2, 0.2, 0.2)  # Dark gray text
context.show_text(dev_text)

# Save the logo
surface.write_to_png("jeremy_logo.png")
ipd.Image("jeremy_logo.png")
```
