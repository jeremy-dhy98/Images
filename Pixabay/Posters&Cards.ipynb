```python
# Making a photo transparent for layering
# Load the image
# "graduation-cake-3960020_1280.jpg",
# Image paths
image_paths = ["bee-8978619_1280.jpg", "graduation-1449488_1280.jpg"]

# Process each image
for image_index, image_path in enumerate(image_paths):
    # Load and convert the image to RGBA
    img = Image.open(image_path).convert("RGBA")
    pixels = img.getdata()

    # Define the color to make transparent (e.g., white)
    background_color = (255, 255, 255)  # RGB value of white

    # Create a new pixel data list
    new_pixels = []
    for pixel in pixels:
        # Make pixels matching the background color transparent
        if pixel[:3] == background_color:
            new_pixels.append((255, 255, 255, 0))  # Fully transparent
        else:
            new_pixels.append(pixel)

    # Update the image with the new pixels
    img.putdata(new_pixels)

    # Save the image with a transparent background
    output_path = f"image_{image_index + 1}.png"
    img.save(output_path, "PNG")
    print(f"Saved {output_path}")

    # Display the processed image
ipd.Image("image_1.png")
```
